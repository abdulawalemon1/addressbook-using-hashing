```python
class HashTable:
    def __init__(self):
        self.max=100
        self.arr=[None for i in range(self.max)]
    def get_hash(self,key):
        h=0
        for char in key:
            h+=ord(char)
        return h% self.max
    def setitem(self,key,value):
        h=self.get_hash(key)
        self.arr[h]=value
    def getitem(self,key):
        h=self.get_hash(key)
        return self.arr[h]
    def delitem(self,key):
        h=self.get_hash(key)
        self.arr[h]=None
```

```python
class HashTable_chain:
    def __init__(self):
        self.max=100
        self.arr=[[] for i in range(self.max)]
    def get_hash(self,key):
        h=0
        for char in key:
            h+=ord(char)
        return h% self.max
    def setitem(self,key,value):
        h=self.get_hash(key)
        found=False
        for index, element in enumerate(self.arr[h]):
            if len(element)==2 and element[0]==key:
                self.arr[h][index]=(key,value)
                found=True
                break
        if not found:
            self.arr[h].append((key,value))
        

        
    def getitem(self,key):
        h=self.get_hash(key)
        for element in self.arr[h]:
            if element[0] == key:
                return element[1:2]
```

```python
class HashTable_chain:
    def __init__(self):
        self.max=6
        self.arr=[[] for i in range(self.max)]
        self.arr_phone_number=[[] for i in range(self.max)]
    def get_hash(self,key):
        i=0
        for char in key:
            i=i+ord(char)
        return i% self.max
    def setitem(self,key,value):
        h=self.get_hash(key)
        found=False
        for index, element in enumerate(self.arr[h]):
            if len(element)==2 and element[0]==key:
                self.arr[h][index]=(key,value)
                found=True
                break
        if not found:
            self.arr[h].append((key,value))
        

        
    def getitem(self,key):
        h=self.get_hash(key)
        for element in self.arr[h]:
            if element[0] == key:
                return element[1:2]
            
    def setitem_phone_number(self,key,value):
        h=self.get_hash(key)
        found=False
        for index, element in enumerate(self.arr_phone_number[h]):
            if len(element)==2 and element[0]==key:
                self.arr_phone_number[h][index]=(key,value)
                found=True
                break
        if not found:
            self.arr_phone_number[h].append((key,value))
        

        
    def getitem_phone_number(self,key):
        h=self.get_hash(key)
        for element in self.arr_phone_number[h]:
            if element[0] == key:
                return element[1:2]
```
